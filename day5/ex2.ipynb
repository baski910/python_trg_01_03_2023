```python
# thresholding
# gimp - stadard for weight = [0.299 , 0.587, 0.114]
# [124,100,128] * [0.299 , 0.587, 0.114] = [(124*0.299)+(100*0.587)+(128*0.114)]
gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
img = gray(pic)
plt.figure( figsize = (4,3))
plt.imshow(img, cmap = plt.get_cmap(name = 'gray'))
plt.show()
```
